```julia

mutable struct CalcBuffer
    alpha_n::Array{Float64,1}
    beta_n::Array{Float64,1}
    kappa_n::Array{Float64,1}
    mu_n::Array{Float64,1}
    x_sum::Array{Float64,1}
    x2_sum::Array{Float64,1}
    p_r_x::Array{Float64,1}
    p_r::Array{Float64,2}
    map_p_r::Array{Float64,1}
    
end

```

```julia
function CalcBuffer(n::Int,a::Float64,b::Float64,k::Float64,mu::Float64)  
    CalcBuffer(
        [a],
        [b],
        [k],
        [mu],
        [0.0],
        [0.0],
        [1.0],
        zeros(Float64,n+1,n),
        Array{Float64,1}(undef,n)
    )
end
CalcBuffer(200,1.0,1.0,1.0,0.0)
```

```julia
alpha_0 = 1.0
beta_0  = 1.0
kappa_0 = 1.0
mu_0    = 0.0
l = 200.0   
n = length(x)
t=collect(range(1.0,stop=n,length=n))
buf = CalcBuffer(alpha_0,beta_0,kappa_0,mu_0)

for i in eachindex(x)
    dist = LocationScale.(buf.mu_n, sqrt.(buf.beta_n .* (buf.kappa_n .+ 1.0) ./ (buf.alpha_n .* buf.kappa_n)),TDist.(2.0 .* buf.alpha_n))
    p_x = pdf.(dist,x[i])

    # compute the growth probabilities p(r_t != 0, x_{1:t})
    p_rx_x = (1.0 - 1.0/l) .* p_x .* buf.p_r_x

    # compute the changepoint probability, p(r_t = 0, x_{1:t})
    p_r0_x = (1.0 / l).* dot(p_x, buf.p_r_x)

    # update the probability distribution p(r_t, x_{1:t}) and normalize it to obtain
    # p(r_t|x_{1:t})
    tmp = vcat(p_r0_x, p_rx_x)
    buf.p_r_x = tmp ./ sum(tmp)

    # keep the result in memory
    buf.p_r[1:i+1, i] = buf.p_r_x    # p(r_t|x_{1:t})
    buf.map_p_r[i] = argmax(buf.p_r_x)    # argmax r_t p(r_t|x_{1:t})


    # update sufficient statistics
    buf.x_sum=vcat(0.0,buf.x_sum .+ x[i])
    buf.x2_sum=vcat(0.0,buf.x2_sum .+ (x[i]^2))
    buf.kappa_n = vcat(kappa_0,buf.kappa_n .+ 1.0)    
    buf.alpha_n = vcat(alpha_0, buf.alpha_n .+ 0.5)
    buf.mu_n = ( (kappa_0 * mu_0) .+ buf.x_sum) ./ buf.kappa_n
    buf.beta_n = beta_0 .+ 0.5 .* (mu_0.^2 .* kappa_0 .+ buf.x2_sum .- buf.mu_n .^ 2 .* buf.kappa_n)
end

Plots.scatter(t,x,markersize=2, c=:blue,xaxis=("t", (-2, 200), 0:10:200))
plot!(t,buf.map_p_r)

```
